```python
import pandas as pd
```

```python
dataset_1 = pd.read_csv(r"C:\Users\MIQDAD\data\googleplaystore.csv")
```

```python
print(dataset_1)
```

# cleaning the installs column and converting to int 


```python
dataset_1['Installs'] = dataset_1['Installs'].str.replace('+', '')
dataset_1['Installs'] = dataset_1['Installs'].str.replace(',', '')
dataset_1['Installs'] = dataset_1['Installs'].apply(lambda x: 0 if x == 'Free' else x)
dataset_1['Installs'] = dataset_1['Installs'].astype(int)
```

# the top 10 installed apps 

```python
top_10_apps = dataset_1.nlargest(10, 'Installs')
```

```python
top_10_apps.plot(kind='bar', x='App', y='Installs')
```

```python
print(top_10_apps)
```

# avg of apps and the most installed categories 

```python
mean_installs_by_category = dataset_1.groupby('Category')['Installs'].mean()
```

```python
mean_installs_by_category_sorted = mean_installs_by_category.sort_values(ascending=False)
```

```python
mean_installs_by_category_sorted.plot(kind='bar')
```

```python
print(mean_installs_by_category_sorted)
```
